## Data Correction 

```python
df.info()
```

```python
# Replace "Three" with "3" in the 'member_in_family' column
df['member_in_family'] = df['member_in_family'].replace('Three', '3')
#Convert this column to numerical
df['member_in_family'] = df['member_in_family'].astype(int)

#Replace No with 0 and Yes with 1 in the column 'following_company_page'
df['following_company_page'] = df['following_company_page'].replace('No', '0')
df['following_company_page'] = df['following_company_page'].replace('Yes', '1')

#Replace anything other than 0 by 1 in the column 'Adult_flag'
df['Adult_flag']=df['Adult_flag'].replace(2,1)
df['Adult_flag']=df['Adult_flag'].replace(3,1)

#Correcting the data in the column preferred_device
df['preferred_device']=df['preferred_device'].replace('Android OS','Android')
df['preferred_device']=df['preferred_device'].replace('ANDROID','Android')
df['preferred_device']=df['preferred_device'].replace('Other','Others')


#Correcting the values in preferred_location_type column
df['preferred_location_type']=df['preferred_location_type'].replace('Tour  Travel','Tour and Travel')
df['preferred_location_type']=df['preferred_location_type'].replace('Game','Other')
df['preferred_location_type']=df['preferred_location_type'].replace('OTT','Other')
df['preferred_location_type']=df['preferred_location_type'].replace('Movie','Other')

#Replace No with 0 and Yes with 1 in the column 'working_flag'
df['working_flag'] = df['working_flag'].replace('No', '0')
df['working_flag'] = df['working_flag'].replace('Yes', '1')

#Replace No with 0 and Yes with 1 in the column 'Taken_product'
df['Taken_product'] = df['Taken_product'].replace('No', '0')
df['Taken_product'] = df['Taken_product'].replace('Yes', '1')

#Convert this column to numerical
df['yearly_avg_Outstation_checkins'] = df['yearly_avg_Outstation_checkins'].astype(float)
```

## Outlier Treatment

We are taking the 95th percentile and 5th percentile as the threshold for treating outliers

```python
outliers = ['Yearly_avg_view_on_travel_page',
            'total_likes_on_outstation_checkin_given',
            'member_in_family',
            'Yearly_avg_comment_on_travel_page',
            'total_likes_on_outofstation_checkin_received',
            'montly_avg_comment_on_company_page',
            'Daily_Avg_mins_spend_on_traveling_page']

# Loop through the columns with outliers
for col in outliers:
    # Calculate the 5th and 95th percentiles
    lower_bound = df[col].quantile(0.05)
    upper_bound = df[col].quantile(0.95)

    # Replace outliers with values within the bounds
    df[col] = df[col].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))

```

# Clustering

```python
#first remove the target variable from the data to create a data set for clustering. 
df_clus=df_encoded.drop('Taken_product',axis=1)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
silhouette_scores = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=0)
    labels = kmeans.fit_predict(df_clus)  # Cluster labels for each data point
    silhouette_avg = silhouette_score(df_clus, labels)
    silhouette_scores.append(silhouette_avg)
```

```python
plt.plot(range(2,10), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different Numbers of Clusters')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Define the number of clusters (n)
n_clusters = 2

# Initialize the KMeans model
kmeans = KMeans(n_clusters=n_clusters, random_state=0)

# Fit the model to the data and obtain cluster labels
cluster_labels = kmeans.fit_predict(df_clus)

# Add the cluster labels to the DataFrame
df_clus['Cluster'] = cluster_labels
```
